```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
records = pd.read_csv('../data/processed/records.csv')
records['created_at'] = pd.to_datetime(records['created_at'])
```

```python
# DÃ­a random
start_time = pd.Timestamp('2024-07-03 00:00', tz='UTC')
end_time = pd.Timestamp('2024-07-03 23:59', tz='UTC')

# Filtrar el DataFrame
day_records = records.loc[(records['created_at'] >= start_time) & (records['created_at'] <= end_time)]
```

```python
# Dummy plot
plt.figure(figsize=(10, 6))
plt.plot(day_records['created_at'], day_records['sgv'], marker='o')
plt.scatter(day_records['created_at'], day_records['carbs'], marker='s')
plt.scatter(day_records['created_at'], day_records['insulin'], marker='v')
plt.scatter(day_records['created_at'], day_records['duration'], marker='P')
plt.xlabel('Timestamp')
plt.ylabel('SGV')
plt.title('Day records plot')

# Formatear el eje x para mostrar solo las horas
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```
